```python
np.random.seed(0)
df4 = pd.DataFrame(np.round(np.random.randn(6,4),2),
                   columns=[['A','A','B','B'],
                            ['C1','C2','C1','C2']],
                  index=[['M','M','M','F','F','F'],
                        ['id_'+ str(i+1) for i in range(3)] *2])
df4.columns.names = ["cidx1","cidx2"]
df4.index.names = ['Ridx1','Ridx2']
df4
```

```python
df4.unstack('Ridx2')
```

date_range파라미터 링크 = pandas.pydata.org/pandas-docs/stable/user_guide/timeseries.html#dateoffset-objects  
● s: 초  
● T: 분  
● H: 시간  
● D: 일(day)  
● B: 주말이 아닌 평일  
● W: 주(일요일)  
● W-MON: 주(월요일)  등등

```python
#연습문제
import pandas as pd
import seaborn as sns
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
def change_boolean(value):
    if value == 'yes':
        return True
    elif value == 'no':
        return False
titanic.alive.apply(change_boolean)
```

```python
titanic.dropna(subset = 'age',inplace=True)
titanic.age.isnull().sum()
```

```python
titanic.age
```

```python
titanic
```

```python
#연습문제
titanic['adult_femail'] = titanic.apply(lambda x : True if (x.age >= 20) & (x.sex =='femail') else False , axis=1)
titanic['adult_femail']
```

```python
titanic.reset_index().to_csv('titanic_age.csv')
```
